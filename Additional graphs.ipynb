```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
X = np.linspace(0,1,1000)

def f(x):
    return (0.6*(5*(x-0.5)**3 + 0.7) + 0.2*x)/1.15
def f1(x):
    return (0.6*(5*(x-0.5)**3 + 0.5) + 0.22*x)/1.15
def f2(x):
    return (0.75*(5*(x-0.55)**3 + 1) + 0.15*x)/1.2
k1 = np.zeros(1000)
k1 = k1+0.5
k2 = k1 + 0.1

```

```python
plt.figure(figsize=(7,6))
plt.plot(X,f(X), label = "$y = f_R(T_x)$")
plt.xlabel("$T_x$", fontsize = 20)
plt.ylabel("$y$", fontsize = 20)
plt.legend(fontsize = 20)
plt.savefig("g1.jpg")
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.plot(X,f2(X), label = "$y = f_{R_A}(T_x)$")
plt.plot(X,f(X), label = "$y = f_{R_B}(T_x)$")
plt.plot(X,f1(X), label = "$y = f_{R_C}(T_x)$")
plt.plot(X,k1, color = "k")
plt.xlabel("$T_x$", fontsize = 20)
plt.ylabel("$y$", fontsize = 20)
plt.ylim([0,1])
plt.legend(fontsize = 12)
plt.text(0,0.52,"$k_1$",fontsize = 16)
plt.savefig("g2.jpg")
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.plot(X,f(X), label = "$y = f_{R_1}(T_x)$")
plt.plot(X,k1, color = "k")
plt.xlabel("$T_x$", fontsize = 20)
plt.ylabel("$y$", fontsize = 20)
plt.ylim([0,1])
plt.legend(fontsize = 12)
plt.text(0,0.52,"$k_1$",fontsize = 16)
plt.savefig("g3.jpg")
plt.show()
```

```python
plt.figure(figsize=(7,6))
plt.plot(X,f(X), label = "$y = f_{R_1}(T_x)$")
plt.plot(X,f1(X), label = "$y = f_{R_2}(T_x)$")
plt.plot(X,k1, color = "k")
plt.plot(X,k2, color = "k", linestyle='dashed')
plt.xlabel("$T_x$", fontsize = 20)
plt.ylabel("$y$", fontsize = 20)
plt.ylim([0,1])
plt.legend(fontsize = 12)
plt.text(0,0.52,"$k_1$",fontsize = 16)
plt.text(0,0.62,"$k_2$",fontsize = 16)
plt.savefig("g4.jpg")
plt.show()
```

```python
a1 = 0.675
a2 = 0.905
splitA = np.linspace(0,a1,100)
splitB = np.linspace(a1,a2,100)
splitC = np.linspace(a2,1,100)
K1SA = np.zeros(100)
K1SA = K1SA + 0.5
K2SA = K1SA + 0.1

def upper(x):
    a = f(x)
    for i in np.where(f(x)<0.6):
        a[i]=0.6
    return a
def lower(x):
    a = f1(x)
    for i in np.where(f1(x)>0.5):
        a[i]=0.5
    return a

plt.figure(figsize=(7,6))
plt.plot(X,f(X), label = "$y = f_{R_1}(T_x)$", color = "royalblue")
plt.plot(X,f1(X), label = "$y = f_{R_2}(T_x)$", color = "lawngreen")
plt.plot(X,k1,color = "k")
plt.plot(X,k2,color = "k", linestyle='dashed')
plt.xlabel("$T_x$", fontsize = 20)
plt.ylabel("$y$", fontsize = 20)
plt.ylim([0,1])
plt.fill_between(splitA, f1(splitA), y2=K2SA, color = "grey", alpha = 0.3)
plt.fill_between(splitC, K1SA, y2=f(splitC), color = "grey", alpha = 0.3)
plt.fill_between(splitB, lower(splitB), y2=upper(splitB), color = "coral", alpha = 0.3)
plt.legend(fontsize = 12)
plt.text(0.3,0.53,"$A$", fontsize = 20)
plt.text(0.07,0.37,"$A_1$", fontsize = 20)
plt.text(0.4,0.37,"$A_2$", fontsize = 20)
plt.text(0.93,0.53,"$B$", fontsize = 20)
plt.text(0.91,0.77,"$B_1$", fontsize = 20)
plt.text(0.95,0.63,"$B_2$", fontsize = 20)
plt.text(0,0.52,"$k_1$",fontsize = 16)
plt.text(0,0.62,"$k_2$",fontsize = 16)

plt.savefig("g5.jpg")
plt.show()
```

```python
def sol(x):
    return x**2
def c1(x):
    return x**2 - 0.05
def c2(x):
    return x**2 - 0.2
xc1 = np.linspace(0,0.3873,100)
xc2 = np.linspace(0,0.5477,100)

k1 = np.zeros(100)
k2 = k1 +0.1

c1s1 = np.linspace(0,0.2236,100)
c1s2 = np.linspace(0.2236,0.3162,100)
c1s3 = np.linspace(0.3162,0.3873,100)

c2s1 = np.linspace(0,0.3162,100)
c2s2 = np.linspace(0.3162,0.4472,100)
c2s3 = np.linspace(0.4472,0.5477,100)

fig, axs = plt.subplots(2)
fig.set_figheight(6)
fig.set_figwidth(7)
axs[0].title.set_text('Case 1')
axs[0].title.set_fontsize(20)
axs[1].title.set_text('Case 2')
axs[1].title.set_fontsize(20)
axs[0].set_yticklabels([])
axs[0].set_xticklabels([])
axs[0].set_xticks([])
axs[0].set_yticks([])
axs[1].set_yticklabels([])
axs[1].set_xticklabels([])
axs[1].set_xticks([])
axs[1].set_yticks([])


axs[0].plot(xc1, sol(xc1))
axs[0].plot(xc1, c1(xc1))
axs[0].plot(xc1, k1, color = "k")
axs[0].plot(xc1, k2, color = "k", linestyle='dashed')
axs[0].fill_between(c1s1, c1(c1s1), y2=k2, color = "coral", alpha = 0.3)
axs[0].fill_between(c1s2, k1, y2=k2, color = "coral", alpha = 0.3)
axs[0].fill_between(c1s3, sol(c1s3), y2=k1, color = "coral", alpha = 0.3)
axs[0].text(0.1,0.05,"$A_1$", fontsize = 14)
axs[0].text(0.18,0.01,"$A_2$", fontsize = 14)
axs[0].text(0.02,-0.03,"$A_3$", fontsize = 14)
axs[0].text(0.23,0.074,"$B_1$", fontsize = 14)
axs[0].text(0.26,0.035,"$B_2$", fontsize = 14)
axs[0].text(0.29,0.01,"$B_3$", fontsize = 14)
axs[0].text(0.36,0.11,"$C_1$", fontsize = 14)
axs[0].text(0.325,0.075,"$C_2$", fontsize = 14)
axs[0].text(0.35,0.034,"$C_3$", fontsize = 14)

axs[1].plot(xc2, sol(xc2))
axs[1].plot(xc2, c2(xc2))
axs[1].plot(xc2, k1, color = "k")
axs[1].plot(xc2, k2, color = "k", linestyle='dashed')
axs[1].fill_between(c2s1, c2(c2s1), y2=k2, color = "coral", alpha = 0.3)
axs[1].fill_between(c2s2, c2(c2s2), y2=sol(c2s2), color = "coral", alpha = 0.3)
axs[1].fill_between(c2s3, sol(c2s3), y2=k1, color = "coral", alpha = 0.3)
axs[1].text(0.1,0.05,"$a_1$", fontsize = 14)
axs[1].text(0.27,0.02,"$a_2$", fontsize = 14)
axs[1].text(0.1,-0.1,"$a_3$", fontsize = 14)
axs[1].text(0.42,0.13,"$b_1$", fontsize = 14)
axs[1].text(0.37,0.03,"$b_2$", fontsize = 14)
axs[1].text(0.33,-0.05,"$b_3$", fontsize = 14)
axs[1].text(0.5,0.2,"$c_1$", fontsize = 14)
axs[1].text(0.46,0.05,"$c_2$", fontsize = 14)
axs[1].text(0.52,0.035,"$c_3$", fontsize = 14)

plt.savefig("g6.jpg")
plt.show()
```

```python
from scipy.stats import beta
```

```python
plt.figure(figsize=(10,5))
a, b = 2, 5
x = np.linspace(beta.ppf(0, a, b),beta.ppf(1, a, b), 100)
plt.plot(x, beta.pdf(x, a, b),'r-', lw=3, alpha=0.6, label='beta(2,5)')
a, b = 5, 2
x = np.linspace(beta.ppf(0, a, b),beta.ppf(1, a, b), 100)
plt.plot(x, beta.pdf(x, a, b),'b-', lw=3, alpha=0.6, label='beta(5,2)')
a, b = 5, 5
x = np.linspace(beta.ppf(0, a, b),beta.ppf(1, a, b), 100)
plt.plot(x, beta.pdf(x, a, b),'g-', lw=3, alpha=0.6, label='beta(5,5)')
plt.xlabel("$T_x$", fontsize = 15)
plt.ylabel("PDF", fontsize = 15)
plt.legend(fontsize = 15)
plt.title("PDF of three different beta distributions", fontsize=15)
plt.plot()
```
